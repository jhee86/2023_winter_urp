## 형태소 명사 추출

```python
#형태소분석기 호출★
kiwi = Kiwi()
```

```python
 #명사 추출 함수 (kiwi 0.10.0)
def get_noun(texts):
    res = kiwi.analyze(texts)[0][0]
    return [Token[0]
            for Token in res
            if Token[1]=='NNG' or Token[1]=='NNP']

#함수 동작 체크
get_noun('금잔디에 가면 볶음우동을 먹을 수 있어요')


# 최초 형태소 분석★
# 전처리 후 다시 형태소분석을 진행할 때도 이 셀을 실행해주세요.
given_data['nouns'] = given_data['제목'].progress_map(lambda x:get_noun(x))
```

```python
class CustomTokenizer:
    def __init__(self, tagger):
        self.tagger = tagger
    def __call__(self, sent):
        sent = sent[:1000000]
        #word_tokens = self.tagger.morphs(sent)
        #result = [word for word in word_tokens if len(word) > 1]
        result = [word[0] for word in self.tagger.analyze(sent, top_n=1)[0][0]if (word[1]=='NNG' or word[1]=='NNP') and (len(word[0]) >1)]
        return result
```

```python
#custom_tokenizer = CustomTokenizer(Mecab())
custom_tokenizer = CustomTokenizer(kiwi)
```

```python
# 토픽에 해당하는 문서를 분류하고 분류된 문서 수를
#-1번토픽은 어떠한 것에 할당되지 않음

# 각 토픽에 해당하는 문서를 분류하고 분류된 문서 수를 저장할 딕셔너리를 초기화
topic_articles = {}


# 토픽 결과를 하나씩 확인
for i in range(len(topics)):
  # 토픽 번호를 가져옵니다.
    topic_num = topics[i]
     # 만약 토픽이 할당된 것이라면 (False가 아니라면)
    if topic_num is not False:
      # 토픽 번호를 정수로 변환합니다.
        topic_num = int(topic_num)
         # 해당 토픽 번호에 대한 리스트가 딕셔너리에 없다면 생성
        if topic_num not in topic_articles:
            topic_articles[topic_num] = []
            # 해당 토픽에 할당된 문서 번호를 리스트에 추가
        topic_articles[topic_num].append(i)

# 분류된 문서 수 출력
for topic_num, articles in topic_articles.items():
    print(f"{topic_num}번 토픽: {len(articles)}개 문서")
```

```python
preprocessed_documents=given_data['nouns'].tolist()
```

```python
# 토픽에 할당된 문서를 분류하고 분류된 문서 수와 함께 문서 내용을 출력합니다.
for topic_num, articles in topic_articles.items():
    print(f"{topic_num}번 토픽: {len(articles)}개 문서")
    for article_index in articles:
        print(f"  - 문서 번호: {article_index}")
        print(f"    내용: {preprocessed_documents[article_index]}")
        print()
```

```python
# 파일에 쓰기를 수행하는 함수
def write_results_to_file(output_file_path, topic_articles, preprocessed_documents,model):
    with open(output_file_path, "w", encoding="utf-8") as file:
        for topic_num, articles in topic_articles.items():
            file.write(f"{topic_num}번 토픽:{model.get_topic(topic_num)} \n{len(articles)}개 문서\n")
            for article_index in articles:
                file.write(f"  - 문서 번호: {article_index}\n")
                file.write(f"    내용: {preprocessed_documents[article_index]}\n\n")

# 예시로 파일 경로를 "output_topic_results.txt"로 지정
output_file_path = "output_topic_results.txt"

# 함수를 호출하여 결과를 파일에 쓰기
write_results_to_file(output_file_path, topic_articles, preprocessed_documents,model)

print(f"Results saved to: {output_file_path}")
```
