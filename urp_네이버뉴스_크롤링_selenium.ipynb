```python
#함수정의
# 페이지 url 형식에 맞게 바꾸어 주는 함수 만들기
# 입력된 수를 1, 11, 21, 31 ...만들어 주는 함수
def makePgNum(num):
    if num == 1:
        return num
    elif num == 0:
        return num + 1
    else:
        return num + 9 * (num - 1)


# 크롤링할 url 생성하는 함수 만들기(검색어, 크롤링 시작 페이지, 크롤링 종료 페이지)

def makeUrl(search, start_pg, end_pg):
    import random
    num_random_pages = 100 #(랜덤으로 100페이지 검색어당 1000개의 뉴스기사)
    seed_value = 42  # 원하는 시드 값을 사용하세요.
    random_pages = random.sample(range(start_pg, end_pg + 1), num_random_pages)


    if start_pg == end_pg:
        start_page = makePgNum(start_pg)
        url = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query=" + search + "&start=" + str(
            start_page)
        print("생성url: ", url)
        return url
    else:
        urls = []
        for i in random_pages:
            page = makePgNum(i)
            url = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query=" + search + "&start=" + str(page)
            urls.append(url)
        print("생성url: ", urls)
        return urls
```

```python

# 뉴스 내용 크롤링
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.366"}
index=0
print("크롤링시작 전\n")
for i in tqdm(unique_urls[:]):
    print("i:",i)
    #각 기사 html get하기
    news = requests.get(i,headers=headers)
    news_html = BeautifulSoup(news.text,"html.parser")

    # 뉴스 제목 가져오기
    title = news_html.select_one("#ct > div.media_end_head.go_trans > div.media_end_head_title > h2")
    if title == None:
        title = news_html.select_one("#content > div.end_ct > div > h2")


    # html태그제거 및 텍스트 다듬기
    pattern1 = '<[^>]*>'
    title = re.sub(pattern=pattern1, repl='', string=str(title))

    pattern2 = """[\n\n\n\n\n// flash 오류를 우회하기 위한 함수 추가\nfunction _flash_removeCallback() {}"""
    title = title.replace(pattern2, '')




#날짜
    try:
        html_date = news_html.select_one("div#ct> div.media_end_head.go_trans > div.media_end_head_info.nv_notrans > div.media_end_head_info_datestamp > div > span")
        news_date = html_date.attrs['data-date-time']
    except AttributeError:
        news_date = news_html.select_one("#content > div.end_ct > div > div.article_info > span > em")
    news_date = re.sub(pattern=pattern1,repl='',string=str(news_date))





#언론사
    try:
        html_mediaC = news_html.select_one("#ct > div.media_end_head.go_trans > div.media_end_head_top._LAZY_LOADING_WRAP > a > img.media_end_head_top_logo_img.light_type._LAZY_LOADING._LAZY_LOADING_INIT_HIDE")

        news_mediaC = html_mediaC.attrs['title']
    except AttributeError:
        news_mediaCs =None
        pass
    news_mediaC = re.sub(pattern=pattern1,repl='',string=str(news_mediaC))





#댓글수- 셀레늄으로 크롤링
    driver.get(i)
    time.sleep(3)

    try:
      commentN_xpath = '//*[@id="cbox_module"]/div[2]/div[1]/a/span[1]'
      commentN_button = driver.find_element(By.XPATH, commentN_xpath)
      commentN = commentN_button.text
    except:
      commentN =None
      pass



    df.at[index, 'link'] = i

    df.at[index, '작성날짜'] = news_date
    df.at[index, '제목'] = title
    df.at[index, '언론사'] = news_mediaC
    df.at[index, '댓글수'] = commentN


    index+=1
    print('index : ', index)

    #now = datetime.datetime.now()
    df.to_csv('/content/drive/MyDrive/2023 겨울계절 URP/데이터_수집/{}_네이버뉴스.csv'.format(issue),encoding='utf-8-sig',index=False)





print("검색된 기사 갯수: 총 ",(page2+1-page)*10,'개')
```
